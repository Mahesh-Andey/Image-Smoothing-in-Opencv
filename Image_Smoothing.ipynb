#**Image Smoothing**

```python
# The function applies an arbitrary linear filter to an image.

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('/content/macbook.jpg')

kernel = np.ones((40,40),np.float32)/1600  # Kernel operation is continued for all the pixels in the image. Here we can predefine kernel

dst = cv.filter2D(img,-1,kernel)

plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Filtering')
plt.xticks([]), plt.yticks([])
plt.show()

# The operation works like this: keep this kernel above a pixel, add all the 1600 pixels below this kernel, take the average, and replace the central pixel with the new average value. 
```

```python

# convolving the image with a low-pass filter kernel
# remove noise
# removes high frequency content (eg: noise, edges) from the image. 

# No predefined kernel techniques are given below :-

###### 1) Averaging 

img = cv.imread('/content/macbook.jpg')

blur = cv.blur(img,(50, 50))
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
######### 2) Gaussian Blurring (Gaussian kernel)

# highly effective in removing Gaussian noise from an image.

blur = cv.GaussianBlur(img,(145, 145),0)  # 1) width and height of the kernel 2) standard deviation in the X and Y directions 
                            # width and height should be positive and odd
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
########### 3. Median Blurring

# takes the median of all the pixels under the kernel area and the central element is replaced with this median value.

# highly effective against salt-and-pepper noise(impulse noise) in an image. 
# Normally if there is black dots in the image we called it pepper noise and if there is white dots in the image we called it salt noise. This noise is generally caused by errors in data transmission, failure in memory cell or analog-to-digital converter errors.

img2 = cv.imread('/content/Salt-Pepper-noise.jpg')
blur = cv.medianBlur(img2, 9)  # kernel size should be a positive odd integer.

plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img2),plt.title('Original with salt and pepper noise')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Median Blurring')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
################ 4. Bilateral Filtering

# highly effective in noise removal while keeping edges sharp.
# slower compared to other filters.
# Bilateral filtering takes a Gaussian filter in space, but one more Gaussian filter which is a function of pixel difference. 

blur = cv.bilateralFilter(img,35,500,500)  # 1-Sigma values - If they are small (< 10), the filter will not have much effect, whereas if they are large (> 150), they will have a very strong effect, making the image look "cartoonish".
                                        # 2-Filter size -  Large filters (d > 5) are very slow, so it is recommended to use d=5 for real-time applications
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bilateral Blurring')
plt.xticks([]), plt.yticks([])
plt.show()
```
